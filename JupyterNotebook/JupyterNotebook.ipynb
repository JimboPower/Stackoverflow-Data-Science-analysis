## Question 1

```python
# Question 1 - Most frequent languages used by professionals in 2019
# Filter the dataframe selecting the columns I need 
q1_19 = df_19[['MainBranch', 'LanguageWorkedWith']]
q1_19 = q1_19.dropna(subset = ['LanguageWorkedWith'], axis = 0)
q1_19 = q1_19[q1_19.MainBranch == 'I am a developer by profession']
languages_19_df = q1_19['LanguageWorkedWith'].str.split(';', expand = True).stack()
languages_19 = languages_19_df.value_counts()

# Creating the graph
q1_19x = (languages_19/q1_19.shape[0]).plot.bar(rot=0)
plt.rcParams['figure.figsize'] = [40, 20]
plt.xticks(rotation='vertical')
plt.rcParams.update({'font.size': 40})
print((languages_19/q1_19.shape[0])[:5])
```

## Evalutation 2019

The most used programming languages in 2019 were: JavaScript (71%), HTML/CSS(64%), SQL (57%), Java(40%), Python (37%).

```python
# Question 1 - Most frequent languages used by professionals in 2018
# Filter the dataframe selecting the columns I need 
q1_18 = df_18[['Employment', 'LanguageWorkedWith']]
q1_18 = q1_18[q1_18.Employment == "Employed full-time"]
q1_18 = q1_18.dropna(subset = ['LanguageWorkedWith'], axis = 0)

# Splitting languages data
languages_18_df = pd.DataFrame(q1_18['LanguageWorkedWith'].str.split(';', expand = True).stack())
languages_18_df[0] = languages_18_df[0].replace({'HTML':'HTML/CSS', "Bash/Shell" : "Bash/Shell/PowerShell"})
languages_18_df = languages_18_df[languages_18_df[0] != "CSS"]
languages_18 = languages_18_df[0].value_counts()
bx = (languages_18/q1_18.shape[0]).plot.bar(rot=0)

# Creating the graph
plt.rcParams['figure.figsize'] = [40, 20]
plt.xticks(rotation='vertical')
plt.rcParams.update({'font.size': 40})
print((languages_18/q1_18.shape[0])[:5])
```

## Evalutation 2018

The most used programming languages in 2019 were: JavaScript (70%), HTML/CSS(67%), SQL (58%), Java(43%), Bash/Shell/PowerShell (40%).

```python
# Question 1 - Comparison between 2019 and 2018
percentage_langueges_18 = languages_18/q1_18.shape[0]
percentage_langueges_19 = languages_19/q1_19.shape[0]

# Concatenating dataframes 
df_comparison_langueges = pd.concat([percentage_langueges_18.combine(percentage_langueges_18,max).rename('2019').fillna(0), 
                           percentage_langueges_19.combine(percentage_langueges_19,max).rename('2018').fillna(0)], axis=1)[:25]

# Creating the graph
ax = df_comparison_langueges.plot.bar(rot=0)
plt.rcParams['figure.figsize'] = [40, 20]
plt.xticks(rotation='vertical')
plt.rcParams.update({'font.size': 40})
```

# Evalutation

Between 2019 and 2018, the four most used programming languages remained the same, but Python took Bash/Shell/PowerShell's place.



## Question 2

```python
# Question 2 - Where programmers come from in 2019
# Filter the dataframe selecting the columns I need 
q2_19 = df_19[['Country']]
q2_19 = q2_19.dropna(subset = ['Country'], axis = 0)
percentage_country_19 = q2_19['Country'].value_counts()[:20]/q2_19.shape[0]
(percentage_country_19).plot(kind='bar')
print(percentage_country_19[:5])
```

## Evalutation 2019

In the 2019, a large part of programmers come from USA (23%), India (10%), Germany (6%), UK (6%) and Canada (3%).

```python
# Question 2 - Where programmers come from in 2018
# Filter the dataframe selecting the columns I need 
q2_18 = df_18[['Country']]
q2_18 = q2_18.dropna(subset = ['Country'], axis = 0)
percentage_country_18 = q2_18['Country'].value_counts()[:20]/q2_18.shape[0]
percentage_country_18.plot(kind = 'bar')
print(percentage_country_18[:5])
```

## Evalutation 2018

In the 2018, a large part of programmers come from USA (20%), India (13%), Germany (6%), UK (6%) and Canada (3%).

```python
# Question 2 - Comparison between 2019 and 2018  
df_comparison_coutries = pd.concat([percentage_country_19.combine(percentage_country_19,max).rename('2019').fillna(0), 
                           percentage_country_18.combine(percentage_country_18,max).rename('2018').fillna(0)], axis=1)[:20]

# Creating the graph
ax = df_comparison_coutries.plot.bar(rot=0)
plt.rcParams['figure.figsize'] = [40, 20]
plt.xticks(rotation='vertical')
plt.rcParams.update({'font.size': 40})
```

# Evalutation

Between 2019 and 2018, programmers' countries did not change a lot, but there was a remarkable decreasement of indian programmers: -%3 and increasement of the USA +3%.

## Question 3

```python
# Question 3 - Are there more male, female or non-binary programmers? in 2019
# Filter the dataframe selecting the columns I need 
q3_19 = df_19[['Gender']]
q3_19 = q3_19.dropna(subset = ['Gender'], axis = 0)
q3_19 = q3_19['Gender'].str.split(';', expand = True).stack()
q3_19 = q3_19.replace({'Man': 'Male', 'Woman': 'Female'})
gender_19 = q3_19.value_counts()
percentage_gender_19 = gender_19/q3_19.shape[0]

# Creating the graph
(gender_19/q3_19.shape[0]).plot(kind = 'bar')
gender_19/q3_19.shape[0]
```

## Evalutation 2019

In 2019 about 91% of Stack Overflow's community is male and the remaining part (Female, Non-binary...) is only 9%.

```python
# Question 3 - Are there more male, female or non-binary programmers? in 2018
# Filter the dataframe selecting the columns I need 
q3_18 = df_18[['Gender']]
q3_18 = q3_18.dropna(subset=['Gender'], axis = 0)
q3_18['Gender'].value_counts()
gender_18 = q3_18['Gender'].str.split(';', expand = True).stack().value_counts()

# Creating the graph
percentage_gender_18 = gender_18/np.sum(q3_18['Gender'].value_counts())
(percentage_gender_18).plot(kind = 'bar')
gender_18/q3_18.shape[0]
```

## Evalutation 2018

In 2018 about 92% of Stack Overflow's community is male and the remaining part (Female, Non-binary...) is only 8%.

```python
# Question 3 - Comparison between genders between 2019 and 2018
df_comparison_gender = pd.concat([percentage_gender_19.combine(percentage_gender_19,max).rename('2019').fillna(0), 
                           percentage_gender_18.combine(percentage_gender_18,max).rename('2018').fillna(0)], axis=1)[:20]

df_comparison_gender.head()

# Creating the graph
ax = df_comparison_gender.plot.bar(rot=0)
plt.rcParams['figure.figsize'] = [40, 20]
plt.xticks(rotation='vertical')
plt.rcParams.update({'font.size': 40})
```

# Evalutation 

Between 2018 and 2019 there was an overall increament of Female, non-binary, trasgender genders, but the percentage of the male programmers community remains the heighest.
